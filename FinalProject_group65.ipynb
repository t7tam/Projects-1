# Data Cleaning

Since our topic required three separate datasets that are independent from one another in terms of the data themselves, we chose to clean them up individually and use them instead of a collective one. For the first step in cleaning, we created multiple lists that contain all the base information that will be used to sort the information in the three datasets. The first list is the movie_list, which contains all the movies that we are looking into ordered by the year. The rest of the lists, labeled religion, sexual, bodily, and social, contain the curse words found in the swear words dataset which we organized into their corresponding lists. For example, “dick” would go into the bodily category list. Next, we went into cleaning the datasets. 
For the swear words dataset, we cleaned it by fixing any spelling errors of the movie titles and dropping unnecessary information, such as the time and type columns. We then searched through the curse words and assigned them a category based on which category list the curse word was in. We then organized the dataset in order of movie based on the movie_list.
For the imdb dataset, we dropped all the unnecessary information by dropping all rows that didn’t have Quentin Tarantino as the director, and specifically the row for the movie, “The Hateful Eight” because it is out of our time range of Quentin Tarantino’s movies we are analyzing. We then isolated the columns to be movie title, gross, budget, imdb score, and year. The use of the year column was to lastly organize the data into the same order as the movie_list.
For the awards dataset, because there were issues with the film title and director name having either an extra space in the text and/or being flipped. Our first step was to get rid of the extra space by stripping the name and the movie title. We then isolated only the rows where the movie title or movie director name matched with the movies in movie_list. Afterwards, because of the issue where the movie title was flipped with the movie director name, we made a for loop that would check to see if the movie title was in the director name column, and then swapped them. The last step was to isolate the dataset even more into just the winner and movie name column.

```python
#Movie and Category Lists
movie_list=['Reservoir Dogs', 'Pulp Fiction', 'Jackie Brown', 'Kill Bill: Vol. 1', 'Kill Bill: Vol. 2', 'Inglourious Basterds', 'Django Unchained']
religion=['goddamned', 'bastards', 'bastard', 'goddamn', 'damn', 'damned', 'hell']
sexual=['fucked', 'fucking', 'fuck', 'fucks', 'fucker', 'fuckers', 'fuckup', 'motherfucker',  'cockblockery',  'fuckface', 'cocksucker', 'fuckhead', 'motherfuckers', 'motherfucking']
body=['dick', 'dicks', 'dumbass','jackass', 'shit', 'shitload', 'horseshit',  'shittiest', 'pussy', 'bullshit', 'shitless', 'shitting', 'horeshit', 'merde', 'shithead',  'chickenshit', 'dickless', 'ass', 'asses', 'asshole', 'assholes', 'asshead', 'shitty']
social=['slut', 'squaw', 'negro', 'negro ', 'jew', 'jew (verb)', 'n-word','n-word ', 'gook', 'gooks', 'slope', 'bitch', 'bitches', 'faggot', 'jap', 'japs', 'wetback', 'cunt', 'cunts',]

#Swear Words Dataset Clean Up
words = pd.read_csv('tarantino.csv')
words['movie'] = words['movie'].replace('Inglorious Basterds', 'Inglourious Basterds')
words = words.drop(['minutes_in', 'type'], axis=1)
words = words.dropna()
words['category'] = np.nan
words.loc[words['word'].isin(religion), 'category'] = 'religion'
words.loc[words['word'].isin(sexual), 'category'] = 'sexual'
words.loc[words['word'].isin(body), 'category'] = 'bodily'
words.loc[words['word'].isin(social), 'category'] = 'social'
sorterIndex = dict(zip(movie_list,range(len(movie_list))))
words['movie_rank'] = words['movie'].map(sorterIndex)
words.sort_values(['movie_rank'], ascending = [True], inplace = True)
words = words.drop(['movie_rank'], axis=1)
print(words['category'].isnull().sum())
print(words)

#IMDB Dataset Clean Up
imdb = pd.read_csv('movie_metadata.csv')
imdb = imdb.drop(imdb[imdb.director_name != 'Quentin Tarantino'].index)
imdb = imdb.drop(706)
imdb = imdb.loc[:,['movie_title', 'gross', 'budget', 'imdb_score', 'title_year']]
imdb = imdb.sort_values(['title_year'])
print(imdb)

#Awards Dataset Clean Up
awards = pd.read_csv('database.csv')
awards['Film'] = awards['Film'].str.strip()
awards['Name'] = awards['Name'].str.strip()
awards = awards[awards['Film'].isin(movie_list) | awards['Name'].isin(movie_list)]
awards = awards.loc[:,['Winner', 'Film', 'Name']]
for i,award_row in awards.iterrows():
    for movie_index in range(0, len(movie_list)):
        name_storage = awards.loc[i,'Name']
        if name_storage == movie_list[movie_index]:
            film_storage = award_row["Name"]
            awards.loc[i,'Name'] = award_row['Film']
            awards.loc[i,'Film'] = film_storage
awards = awards.loc[:,['Winner', 'Film']]
print(awards)
```

# Ethics & Privacy

As big data plays an increasingly crucial role in our everyday lives, ethical and privacy considerations regarding dataset access, biases, misuse, and personal information breaches become more necessary. Given the serious consequences of such violations, we proceeded with our project with ethical and privacy considerations every step of the way. We must note the mature content of our project and obscenities found in the datasets as a result of the vulgar nature of Quentin Tarantino’s films, which include instances of sexism, racism, and graphic violence. The datasets we used are all open, authorized for public use, and accessible by everyone as public resources. Our first dataset, containing entries of all curse words, deaths, and timestamps of when they appear in Tarantino's movies--with the exception of Death Proof, was published by a user on GitHub and subsequently featured on “FiveThirtyEight”, an open-access, dataset-publishing website. Our second and third datasets are from the website “Kaggle”, a popular online public community for finding and publishing open datasets. We did not have to deal with any privacy or terms of use concerns as they do not infringe on any individual’s privacy since all information collected is public and specifically about films. An issue we faced regarding potential biases was how we chose what data to collect and display, as some of the datasets we used had missing information, and we subsequently had to drop two films. In this case, we had to exclude the films Death Proof and The Hateful Eight, as the “FiveThirtyEight” dataset did not include these films. As a result, an ethical consideration is that there will be bias in the data collection since it does not contain all of Quentin Tarantino’s films and that the analysis may not be equitable. 

There is also potential bias and error in the collection of data for the datasets themselves: the creator of the “FiveThirtyEight” dataset notes that, since he personally went through each movie and manually kept track of the curse word count, “it [was] occasionally difficult to make out the profane language precisely” but that he did his “best to count all the curses.” Additionally, an ethical consideration that may result in some bias is that as we are comparing the categories of swear words to the success of Tarantino’s films, we will have to define what we consider to be “successful” and that definition may conflict with what others consider to be successful. Another potential area of subjectivity is the way in which we defined and categorized the swear words. We used Cognitive Science professor Benjamin Bergen’s book What the F: What Swearing Reveals About Our Language, Our Brains, and Ourselves (2016) to guide our categorization, but it is still subjective categorization and anyone could make an argument against our swear word placement in a category. For safe measure, we asked Benjamin Bergen himself where “cunt” should be since it is technically a body organ and is derogatory towards women, and he replied that “cunt” should be in the derogatory section. Our analysis also runs the risk of being used by others to either support or discourage profanity in films and may be used as an argument for films to have more or fewer swear words. We conducted this analysis with no agenda and with a clear reminder that correlation does not equal causation. 


# Conclusion & Discussion

So, is there a relationship between the type of curse words used in Quentin Tarantino’s movies (excluding Death Proof and The Hateful Eight) and the success of those movies? We do note that our research was centered around only seven movies and excludes two important films, so our data is not large enough to reject or support a hypothesis. From our groundwork and analysis, we were not able to find enough significant evidence to confidently say there is a correlation. Also worth mentioning is that we found a scatterplot that demonstrated a strong correlational relationship between bodily function/organ swear words and IMDb scores.
All in all, we found minimal evidence that may support our hypothesis with movies such as Pulp Fiction and Django Unchained, but our sample size is too small to draw any conclusions with confidence and our evidence is not strong enough. Our work may impact how people view swearing in Tarantino’s films and might be used to further an argument for or against swearing, though we doubt people would trust the legitimacy of our research given the limitations we have. The best step forward would be to find a way to include Death Proof and The Hateful Eight into our dataset and observe if it rejects our hypothesis.
